```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(115, size=(115))
y = np.random.randint(115, size=(115))
colors = np.random.randint(115, size=(115))
sizes = 7 * np.random.randint(115, size=(115))

plt.scatter(x,y, c=colors, s=sizes, alpha = 0.4, cmap = 'plasma_r')
                              
plt.colorbar()
                            
plt.show()
```
